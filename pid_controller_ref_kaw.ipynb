```python
import pandas as pd
import matplotlib.pyplot as plt

# read file
file10 = "pid_log_ref1kaw0.csv"
file15 = "pid_log_ref1kaw5.csv"
file30 = "pid_log_ref3kaw0.csv"
file35 = "pid_log_ref3kaw5.csv"

d10 = pd.read_csv(file10)
d15 = pd.read_csv(file15)
d30 = pd.read_csv(file30)
d35 = pd.read_csv(file35)

# --- Plot 1: Output vs Reference  ---
plt.figure()
plt.plot(d10["time"], d10["reference"], linestyle="--", label="Reference1")
plt.plot(d10["time"], d10["output"], label="Ref1, Kaw0)")
plt.plot(d15["time"], d15["output"], label="Ref1, Kaw5")
plt.plot(d30["time"], d30["reference"], linestyle="-", label="Reference5")
plt.plot(d30["time"], d30["output"], label="Ref3, Kaw0)")
plt.plot(d35["time"], d35["output"], label="Ref3, Kaw5")
plt.xlabel("Time [s]")
plt.ylabel("Output")
plt.title("Step Response Comparison")
plt.grid(True)
plt.legend(loc = 'upper right')
plt.show()

# --- Plot 2: Control Input  ---
plt.figure()
plt.plot(d10["time"], d10["control"], label="Ref1, Kaw0")
plt.plot(d15["time"], d15["control"], label="Ref1, Kaw5")
plt.plot(d30["time"], d30["control"], label="Ref3, Kaw0")
plt.plot(d35["time"], d35["control"], label="Ref3, Kaw5")
plt.xlabel("Time [s]")
plt.ylabel("Control Input u")
plt.title("Control Input Comparison")
plt.grid(True)
plt.legend()
plt.show()
```
